```python
import pandas as pd

# 1) Charger le dataset
# Remplace "netflix_titles.csv" par le nom de ton fichier si besoin
df = pd.read_csv("netflix.csv")

# 2) Garder seulement les films
movies = df[df["type"].str.lower() == "movie"].copy()

# 3) Colonnes dans lesquelles chercher des mots-clés apocalypse
search_cols = ["title", "listed_in", "description"]

# 4) Mots-clés liés à l'apocalypse
keywords = [
    "apocalypse", "post-apocalyptic", "post apocalyptic",
    "end of the world", "zombie", "outbreak", "pandemic",
    "virus", "survival", "disaster", "catastrophe",
    "nuclear", "dystopia", "dystopian"
]

# 5) Fusionner le texte des colonnes à analyser
movies["search_text"] = movies[search_cols].fillna("").agg(" ".join, axis=1).str.lower()

# 6) Filtrer les films apocalypse
mask = movies["search_text"].apply(lambda x: any(word in x for word in keywords))
apocalypse_movies = movies[mask].copy()

# 7) Tirer jusqu'à 30 films aléatoires
n = min(30, len(apocalypse_movies))
random_30 = apocalypse_movies.sample(n=n, random_state=None)

# 8) Afficher quelques colonnes utiles
result = random_30[["title", "release_year", "listed_in", "description"]]

print(f"{len(result)} films apocalypse trouvés aléatoirement :")
display(result.reset_index(drop=True))
```
